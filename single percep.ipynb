```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 
df=pd.read_csv("data.csv")
y=df.label
features = ['x1','x2','x3','x4']
X=df[features]
X_train, X_test, y_train, y_test=train_test_split(X,y, random_state=1)
y_train=y_train.to_numpy()
X_train=X_train.to_numpy()
X_test=X_test.to_numpy()
y_test=y_test.to_numpy()
```

```python

class Perceptron(object):
    """Implements a perceptron network"""
    def __init__(self, input_size, lr=0.2,th=0.1, epochs=200):
        self.W = np.zeros(input_size+1)
        self.epochs = epochs
        self.lr = lr
        self.th=th
    
    def activation_fn(self, x):
        
        return 1 if x >=0 else 0
 
    def predict(self, x):
        
        z= np.dot(x, self.W[1:]) + self.W[0]
        a = self.activation_fn(z)
        return a
 
    def fit(self, X, d):
        for _ in range(self.epochs):
            for i in range(d.shape[0]):
                
              
                y = self.predict(X[i])
                e = d[i] - y
                
                self.W[1:] += self.lr * e * X[i]
                self.W[0] += self.lr * e
                #self.th=self.th-(self.lr * e)
        return self.W
        
            
            
           
```

```python
if __name__ == '__main__':

    perceptron = Perceptron(input_size=4)
    perceptron.fit(X_train, y_train)
    print(perceptron.W)
    
    Expected=[]
    Predicted=[]
    for i in range(y_test.shape[0]):
       
        y1 =perceptron.predict(X_test[i])
        Expected=np.append(Expected,y_test[i])
        Predicted=np.append(Predicted,y1)
        print("Expected=%d, Predicted=%d" % (y_test[i],y1))
    
    print(Expected,Predicted)
    results = confusion_matrix(Expected, Predicted)
    print(results)
    print ('Accuracy Score :',accuracy_score(Expected, Predicted)) 
    print ('Report : ')
    print (classification_report(Expected, Predicted) )
    
```
